```python
import pandas as pd
import pyqrcode as pqr
import smtplib 
from email.mime.multipart import MIMEMultipart 
from email.mime.text import MIMEText 
from email.mime.base import MIMEBase 
from email import encoders 

def sendemail(email,fname):
    fromaddr = "debaani20@gmail.com"
    toaddr = email
    msg = MIMEMultipart() 
    #sender email
    msg['From'] = fromaddr 
    #receivers email address 
    msg['To'] = toaddr 
    msg['Subject'] = "Testing"
    body = "Hey Check Your QR Code" 
    msg.attach(MIMEText(body, 'plain')) 
    #file name
    filename = fname
    attachment = open("/home/debasis/Desktop/QRCode/"+fname, "rb") 
    p = MIMEBase('application', 'octet-stream') 
    p.set_payload((attachment).read()) 
    encoders.encode_base64(p) 

    p.add_header('Content-Disposition', "attachment; filename= %s" % filename) 
    msg.attach(p) 
    s = smtplib.SMTP('smtp.gmail.com', 587) 
    s.starttls() 
    s.login(fromaddr, "debasis30") 
    text = msg.as_string() 
    s.sendmail(fromaddr, toaddr, text) 
    s.quit() 

def createQR():
    df=pd.read_csv("scity.csv")
    #print(df.head())
    for index,values in df.iterrows():
        email=values["Email"]
        name=values["Name"]
        reg=values["Registration"]
        course=values["Course"]
        gen=values["Gender"]
        stay=values["Stay"]
        #mob=values["Mobile"]
        
        data=f'''
        Email: {email} \n
        Name: {name} \n
        Registration: {reg} \n
        Course: {course}\n
        Gender: {gen}\n
        Stay: {stay}\n 
        '''
       # print(data)
        fname=f"{name}_{reg}.svg"
        image=pqr.create(data)
        image.svg(fname,scale=1)
        sendemail(email,fname)
createQR()
```
